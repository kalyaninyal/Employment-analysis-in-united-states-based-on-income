# USA Diversity dataset according to county 

```python
diversity_csv = "/Resources/diversityindex.csv"
diversity_df = pd.read_csv(diversity_csv)
diversity_df.head()
```

```python
newdiversity_df = diversity_df
```

```python
newdiversity_df['County'],newdiversity_df['State'] = diversity_df['Location'].str.rsplit(', ').str
newdiversity_df.head()
```

```python
newdiversity_df1 = pd.DataFrame({"Diversity-Index":newdiversity_df.groupby(["State","County"])["Diversity-Index"].mean(),
                       "Black":newdiversity_df.groupby(["State","County"])["Black or African American alone, percent, 2013"].mean(),
                       "American Indian":newdiversity_df.groupby(["State","County"])["American Indian and Alaska Native alone, percent, 2013"].mean(),
                       "Asian":newdiversity_df.groupby(["State","County"])["Asian alone, percent, 2013"].mean(),
                       "Native Hawaiian":newdiversity_df.groupby(["State","County"])["Native Hawaiian and Other Pacific Islander alone, percent,"].mean(),
                       "Two or More Races":newdiversity_df.groupby(["State","County"])["Two or More Races, percent, 2013"].mean(),
                       "Hispanic/ Latino":newdiversity_df.groupby(["State","County"])["Hispanic or Latino, percent, 2013"].mean(),
                       "White":newdiversity_df.groupby(["State","County"])["White alone, not Hispanic or Latino, percent, 2013"].mean()})
```

# The Median income &Population dataset

```python
median_income = pd.read_csv('/Users/vinaybhore/storage/Security/gitrepo/ETL-Project/Resources/2015 Median Income by County.csv')
median_income.head()
```

```python
median_income = median_income[['County',"State Code", 'Population',"Median household income"]]
median_income.head()
```

```python
median_income = median_income.rename(columns={"State Code": "State"})
```

```python
median_income_new = pd.DataFrame({"Median household income":median_income.groupby(["State","County"])["Median household income"].mean(),
                              "Population":median_income.groupby(["State","County"])["Population"].sum()})
```

```python
median_income_new.head()
```

# Merging Datasets

```python
merged = pd.merge(unemployement_df_new,median_income_new, on=["State",'County'],how="inner")
```

```python
merged.head()
```

```python
merged2 = pd.merge(merged,newdiversity_df1, on=["State",'County'],how="inner")
```

```python
merged2.head()
```

```python
merged_final = merged2.reset_index()
```
